# Intro: Linear and Planar Starscapes

This code draws the images for the paper **Algebraic Starscapes** by Edmund Harriss, Kate Stange, and Steve Trettel, which has a lot more detail and discussion of the images. 

The basic idea is to plot families of complex roots of polynomials as dots of different sizes. One of the motivating images is the quadratics so we start with them. 

```python
pts_quad=polyFamily(lambda a,b,c:a*x^2+b*x+c, 25)
```

```python
fig_quad=rootPicture(pts_quad)
```

```python
savePDF(fig_quad, "Quadratics")
```

The family can be created by the function _polyFamily(poly, n)_, where poly defines the polynomials as a lambda function, writing _lambda_ folowed by the parameters you will use (_a,b,c_ for example) then _:_ and the polynomial. For example the quadratics are given by _lambda a,b,c: a*x^2+b*x+c_. 

In general, families of roots with up to 4 parameters are generated, the first parameter will have values from 0 to n, the others from -n to n. 

There are several other families below, some with images and others without (to keep the file size down). The creation of the dataset can take a little time is separated from the drawing so different options can be used. 

An image is then generated by _rootPicture(pts)_.

In general the images are plotted centered on i, so ranging from -1 to 1 for the real part and 0 to 2 for the imaginary. To change this you can set the options xlims and ylims in the rootPicture function (e.g. _xlims=(-2,2), ylims=(0,3)_ to make the real axis go from -2 to 2 and the imaginary axis from 0 to 3).

By default the points are coloured by their degree with black for quadratics, red for cubics, blue for quartics, purple for quintics and then grey beyond that. This can be changed in the dcols list defined above. Points are sized by the root discriminant and drawn as circles. There are more options for colour, size and shape that are explored later. 

You can save a figure using _savePDF(figure, name, \[path\])_ where _figure_ is the figure created above, _name_ is the file name (without extension) and _path_ is the optional file path.

## "The" Curve between algebraic numbers

Studying integer linear combinations of two polynomials gives a curve on the complex plane. For example the roots of the two polynomials *p and q* given below. For polynomials with more than one complex pair of roots this will draw multiple lines as all roots will be plotted.

```python
p = x^3+x^2+2*x+1
q = x^3-x^2-1
pts_line=polyFamily(lambda a,b:a*p+b*q, 20)
fig_line=rootPicture(pts_line)
savePDF(fig_line, "CurveBetweenAlgebraics")
```

The next two images show several of the curves through a single point (*p*). A second polynomial (_q_) gives the other points to draw curves through.   

```python
p = x^3+2*x-1
q = x^3+1
pts_line=polyFamily(lambda a,b,c:a*p+b*(q+c*(x)),30,crange=range(-3,5))
fig_line=rootPicture(pts_line)
savePDF(fig_line, "CurvesBetweenAlgebraicsA0BC")
```

```python
p = x^3-x^2-1
q = x^3+2*x
pts_line=polyFamily(lambda a,b,c:a*p+b*(q+c*(x^2+1)),30,crange=range(-5,3))
fig_line=rootPicture(pts_line)
savePDF(fig_line, "CurvesBetweenAlgebraicsABAC")
```

## Shape

The shape of the dots produced can be changed using the _shapefun_ option. 

Other than the circles used so far there is currently only one option for the shape, designed for cubic roots. Instead of a disc this creates an arrow pointing along the hyperbolic geodesic to the conjugate real root. Note this works for cubic polynomials, so quadratic algebraic numbers appear when the polynomial factors over the integers. For polynomials beyond cubic this will point to whatever conjugate real root is listed first by SAGE. 

```python
pts_abca = polyFamily(lambda a,b,c:a*x^3+b*x^2+c*x+a, 30)
arrows_abca = rootToPt(pts_abca, shapefun=tanBundleArrow)
```
